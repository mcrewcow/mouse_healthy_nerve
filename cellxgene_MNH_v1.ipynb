```python
import mygene
```

```python
fetal_total.var_names
```

```python
mg = mygene.MyGeneInfo()

```

```python
# Query for Ensembl IDs
gene_symbols = fetal_total.var_names.tolist()

gene_symbols
```

```python
query_results = mg.querymany(gene_symbols, scopes='symbol', fields='ensembl.gene', species='mouse')
```

```python

# Convert query results to a DataFrame
df_results = pd.DataFrame(query_results)
```

```python
df_results
```

```python
df_results['ensembl']
```

```python
fetal_total.var
```

```python
fetal_total.var
```

```python
df_results['ensembl_id'] = df_results['ensembl'].apply(lambda x: x['gene'] if isinstance(x, dict) else None)
```

```python
df_results
```

```python
fetal_total.var['ensembl_id'] = df_results['ensembl_id']
```

```python
fetal_total.var
```

```python
gene_symbol_to_ensembl = pd.Series(df_results['ensembl_id'].values, index=df_results['query']).to_dict()
gene_symbol_to_ensembl
```

```python

# Ensure that the dictionary has the correct entries
print("Sample of gene_symbol_to_ensembl mapping:")
for k in list(gene_symbol_to_ensembl)[:10]:
    print(f"{k}: {gene_symbol_to_ensembl[k]}")

```

```python
# Map the Ensembl IDs to the fetal_total.var DataFrame
fetal_total.var['ensembl_id'] = fetal_total.var.index.map(gene_symbol_to_ensembl)
```

```python
fetal_total.var
```

```python
fetal_total.obs['new_donor_id'] = fetal_total.obs.apply(lambda row: f"{row['donor_id']}_{row['GSM']}", axis=1)
```
